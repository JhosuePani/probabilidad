### Ejemplo de cuantiles

Puntuaciones en un Examen

Tienes las puntuaciones de un grupo de estudiantes en un examen. Quieres analizar las medidas de posición para comprender cómo se distribuyen las puntuaciones. Utiliza los conceptos de la mediana, el percentil y el rango intercuartil para responder a algunas preguntas sobre las puntuaciones.

Las puntuaciones de los estudiantes son las siguientes:

```python
import numpy as np

# Puntuaciones de los estudiantes
puntuaciones = [72, 85, 92, 68, 78, 89, 95, 62, 71, 80, 83, 75, 88]
```

Calcula la mediana de las puntuaciones.
1. Encuentra el percentil 75 de las puntuaciones.
2. Calcula el rango intercuartil (Q3 - Q1) de las puntuaciones.
3. Identifica si alguna puntuación es un valor atípico utilizando la regla del rango intercuartil.

```python
# 1. Mediana
mediana = np.median(puntuaciones)

# 2. Percentil 75
percentil_75 = np.percentile(puntuaciones, 75)

# 3. Rango intercuartil (Q3 - Q1)
q1 = np.percentile(puntuaciones, 25)
q3 = np.percentile(puntuaciones, 75)
rango_intercuartil = q3 - q1

# 4. Identificar valores atípicos
valor_min_atipico = q1 - 1.5 * rango_intercuartil
valor_max_atipico = q3 + 1.5 * rango_intercuartil

# Mostrar resultados
print("Mediana:", mediana)
print("Percentil 75:", percentil_75)
print("Rango Intercuartil:", rango_intercuartil)
print("Valores atípicos:", [valor for valor in puntuaciones if valor < valor_min_atipico or valor > valor_max_atipico])
```

### Diagrama de caja y bigotes

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear el boxplot con seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(y=puntuaciones,)
plt.title("Diagrama de Caja de Puntuaciones")
plt.ylabel("Edades")
plt.show()
```

**1. Mediana (línea en el centro de la caja):** La línea en el centro de la caja representa la mediana de las puntuaciones de los estudiantes. En este caso, la mediana está cerca de los 80 puntos. Esto significa que aproximadamente la mitad de las personas tienen puntuaciones menores o iguales a 80 puntos, y la otra mitad tiene puntuaciones mayores o iguales a 80 puntos.

**2. Caja (rango intercuartil):** La caja representa el rango intercuartil (Q1 a Q3), que abarca desde el primer cuartil (Q1) hasta el tercer cuartil (Q3) de las puntuaciones. La altura de la caja muestra la dispersión de las puntuaciones en esta parte central de la distribución.

**3. Bigotes:** Los "bigotes" se extienden desde la caja hasta los valores mínimo y máximo dentro de 1.5 veces el rango intercuartil. Cualquier punto fuera de los "bigotes" se consideraría un valor atípico.

**4. Puntos fuera de los bigotes:** Los puntos individuales que se encuentran fuera de los "bigotes" podrían ser valores atípicos. Sin embargo, en este caso, no hay puntos fuera de los "bigotes", lo que indica que no hay valores atípicos en las puntuaciones.

# Ejemplo de medidas de dispersion

Edades de un Grupo de Personas

Tienes un conjunto de edades de un grupo de personas. Quieres analizar las medidas de dispersión para entender cuánto se dispersan las edades del grupo. Utiliza los conceptos de varianza, desviación estándar, rango y coeficiente de variación para responder a algunas preguntas sobre las edades.

Las edades de las personas son las siguientes:

```python
edades = [25, 30, 22, 27, 40, 35, 28, 32, 29, 26]
```

1. Calcula la varianza de las edades.
2. Calcula la desviación estándar de las edades.
3. Calcula el rango de las edades.
4. Calcula el coeficiente de variación de las edades.

```python
import numpy as np

# 1. Calcula la varianza
media = np.mean(edades)
varianza = np.var(edades, ddof=0)  # ddof=0 para calcular la varianza poblacional

# 2. Calcula la desviación estándar
desviacion_estandar = np.sqrt(varianza)

# 3. Calcula el rango
rango = np.max(edades) - np.min(edades)

# 4. Calcula el coeficiente de variación
coeficiente_variacion = (desviacion_estandar / media) * 100

# Mostrar resultados
print("media:",media)
print("Varianza:", varianza)
print("Desviación Estándar:", desviacion_estandar)
print("Rango:", rango)
print("Coeficiente de Variación:", coeficiente_variacion, "%")
```

1. **Varianza:** La varianza es una medida que indica cuánto se dispersan los valores en relación con la media. En este caso, la varianza de las edades del grupo es aproximadamente 24.4.

2. **Desviación Estándar:** La desviación estándar es la raíz cuadrada de la varianza y es otra medida de dispersión. En este caso, la desviación estándar es aproximadamente 4.94. Indica cuánto se espera que las edades varíen de la media. Valores más altos indican una mayor dispersión.

3. **Rango:** El rango es la diferencia entre el valor máximo y el valor mínimo de los datos. En este caso, el rango de edades es 18. Esto muestra la amplitud total de las edades en el grupo.

4. **Coeficiente de Variación:** El coeficiente de variación es una medida relativa de la dispersión en relación con la media. En este caso, el coeficiente de variación es aproximadamente 16.82%. Esto significa que la desviación estándar es alrededor del 16.82% de la media. Un coeficiente de variación bajo indica que la dispersión es relativamente pequeña en relación con la media.

En resumen, las edades del grupo tienen una dispersión moderada en relación con la media. Aunque las edades varían, la mayoría de ellas tienden a agruparse alrededor de la media, y la dispersión en sí misma no es extremadamente alta.

### Diagrama de caja y bigotes

```python
import seaborn as sns

# Crear el boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=edades)
plt.title("Diagrama de Caja de Edades")
plt.ylabel("Edades")
plt.show()
```

```python
# Crear el gráfico de distribución (histograma)
plt.figure(figsize=(8, 6))
plt.hist(edades, bins=10, edgecolor='black', alpha=0.7)
plt.title("Distribución de edades")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# Ejercicio de asimetría y curtosis
Tienes una lista de ingresos mensuales de un grupo de trabajadores en una empresa. Quieres analizar la asimetría y la curtosis de la distribución de ingresos para comprender si hay una inclinación hacia valores más altos o más bajos. Utiliza el coeficiente de Fisher.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lista de ingresos mensuales en miles de dólares
ingresos = [5, 7, 8, 10, 12, 12, 15, 18, 20, 25, 30, 40, 50]

# Crear un DataFrame con los datos
df = pd.DataFrame({'Ingresos': ingresos})
```

### Coeficiente de asimetría

```python
# Calcular la media y desviación estándar
media = df['Ingresos'].mean()
desviacion_estandar = df['Ingresos'].std()

# Calcular el coeficiente de Fisher
coef_fisher = 3 * (media - df['Ingresos'].median()) / desviacion_estandar

# Interpretación
if coef_fisher < 0:
    interpretacion = "La distribución está sesgada negativamente."
elif coef_fisher > 0:
    interpretacion = "La distribución está sesgada positivamente."
else:
    interpretacion = "La distribución es simétrica."

# Mostrar resultados
print("Coeficiente de Fisher (Asimetría):", coef_fisher)
print(interpretacion)
```

### Coeficiente de curtosis

```python
import pandas as pd

# Calcular el coeficiente de curtosis
coef_curtosis = df['Ingresos'].kurtosis()

# Interpretación
if coef_curtosis > 0:
    interpretacion = "La distribución es leptocúrtica (colas más pesadas y puntiaguda en el centro)."
elif coef_curtosis < 0:
    interpretacion = "La distribución es platicúrtica (colas más ligeras y aplanada en el centro)."
else:
    interpretacion = "La distribución es mesocúrtica (similar a la distribución normal)."

# Mostrar resultados
print("Coeficiente de Curtosis:", coef_curtosis)
print(interpretacion)
```

### Gráfico de distribución

```python
# Crear el gráfico de distribución (histograma)
plt.figure(figsize=(8, 6))
plt.hist(df['Ingresos'], bins=10, edgecolor='black', alpha=0.7)
plt.title("Distribución de Ingresos Mensuales")
plt.xlabel("Ingresos (miles de dólares)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

### Gráfico de densidad de distribución

Un gráfico de densidad de distribución es una representación visual que muestra cómo se distribuyen los datos en una variable continua. 

A diferencia de un histograma, que divide los datos en intervalos y muestra la frecuencia de observaciones en cada intervalo, un gráfico de densidad suaviza la distribución y muestra la "densidad" de los datos en diferentes valores a lo largo del eje horizontal.

Se calcula a partir de la función de densidad de probabilidad y esta depende del tipo de distribución que tengamos.

```python
import seaborn as sns

# Crear el gráfico de densidad de distribución
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df['Ingresos'], fill=True)
plt.title("Densidad de Distribución de Ingresos Mensuales")
plt.xlabel("Ingresos (miles de dólares)")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```
