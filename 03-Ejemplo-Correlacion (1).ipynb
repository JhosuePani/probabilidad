### Matriz de covarianza

```python
# Importar librerias
import numpy as np

# Creamos las listas de datos
X = [10, 5]
Y = [3, 9]
```

```python
# Calcular la varianza
varianza_X = np.var(X, ddof=1)  # Usar ddof=1 para varianza muestral
varianza_Y = np.var(Y, ddof=1)

print(f"Varianza de X: {varianza_X}")
print(f"Varianza de Y: {varianza_Y}")

# Calcular la matriz de covarianza
matriz_cov = np.cov(X, Y, ddof=1)  # Usar ddof=1 para covarianza muestral

print("Matriz de Covarianza:")
print(matriz_cov)

```

```python
# Crear una matriz con los datos X e Y
data = np.array([X, Y])

# Calcular la matriz de correlaci贸n
matriz_correlacion = np.corrcoef(data)

print("Matriz de Correlaci贸n:")
print(matriz_correlacion)
```

### Heatmap

```python
# Para ver la correlacion o los valores de varianzaas
import seaborn as sns
import matplotlib.pyplot as plt

# Crear una matriz de datos de ejemplo
data = np.random.randn(5, 5)

# Calcular la matriz de covarianza y correlaci贸n
covariance_matrix = np.cov(data)
correlation_matrix = np.corrcoef(data)

# Crear heatmaps
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.heatmap(covariance_matrix, annot=True, cmap="coolwarm", fmt=".2f", xticklabels=False, yticklabels=False)
plt.title("Matriz de Covarianza")

plt.subplot(1, 2, 2)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", xticklabels=False, yticklabels=False)
plt.title("Matriz de Correlaci贸n")

plt.show()
```
